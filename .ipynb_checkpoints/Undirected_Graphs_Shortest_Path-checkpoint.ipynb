### Wykorzystanie grafów nieskierowanych do wyznaczenia najkrótszej drogi w danych Open Street Map 

```python
from shapely.geometry import shape
import fiona
import matplotlib.pyplot as plt
import networkx as nx
import geopandas as gpd
import pandas as pd
import contextily as cx
import numpy as np
```

```python
# Zaczytanie plików Open Street Map ze strony https://download.geofabrik.de/
shape_dir = "./dane/drogi_sopot.shp"
geoms = [shape(feature['geometry']) for feature in fiona.open(shape_dir )]
drogi = gpd.read_file(shape_dir)
# Zamiana brakujacych nazw ulic
drogi['name'][drogi['name'].isna()]  = 'bez_nazwy'
```

###  Przekształcenie wektorów dróg do postaci grafów
Przekształcenie odbywa się poprzez przekształcenie geometrycznej reprezentacji geometrii w formacie WKT (par współrzędnych) do postaci połączonych wezłów i łączących je krawędzi.

```python
G = nx.Graph()
n = 0
nn = 0

for line in geoms:
    drogi_s = drogi.iloc[[n]]
    n += 1
    for seg_start, seg_end in zip(list(line.coords),list(line.coords)[1:]):
        nn += 1
        G.add_edge(seg_start,
                   seg_end,
                   N = nn,
                   weight = drogi_s['maxspeed'].values[0], 
                   ulica = drogi_s['name'].values[0],
                   osm_id =  drogi_s['code'].values[0])
```

### Dobór punktu początkowego i końcowego do wyznaczenia najkrótszej trasy
Możemy wybrać dowolną ulicę do wyznaczenia najkrótszej trasy pomiedzy nimi. </br>
Jako obszar zainteresowania wybrano Śródmieście Sopotu. </br>
W przykładzie zaprezentowano drogę od Wydziału Zarządzania UG do Molo. </br>

```python
#Wszystkie nazwy ulic dostępne w obszarze zaintersesowania 
wszystkie_ulice = drogi['name'].unique()
```

```python
def Get_Start_End_Point(Graph, start_point, end_point):
    n = 0
    lbl = []
    W = []
    for node1, node2, data in Graph.edges(data = True):
        W.append(data['weight'])
        lbl.append(data["N"])

        if data["ulica"] == start_point:
            source = node1 
        elif data["ulica"] == end_point:
            target = node1
        n += 1
    return source, target, W, lbl
```

### Wyznaczenie najkrótszej trasy pomiędzy punktami.

```python
def Euclidean_distance(pt_1, pt_2):
    """
    Funkcja wyznaczająca euklidesową odległość
    """
    pt_1 = np.array((pt_1))
    pt_2 = np.array((pt_2))
    return np.linalg.norm(pt_1 - pt_2, axis = 1)

def Get_Path_Attributes(route):    
    i = 0
    list_of_distances = []
    streets_name = []
    speed_limit = []
    for i in range(len(route)- 1):

        data = dict(G.get_edge_data(route[i], route[i + 1]))

        if data['ulica'] not in streets_name:
            streets_name.append(data['ulica'])
            speed_limit.append(data['weight'])

        xy_node1 = np.array(route[i]).reshape((1,2))
        xy_node2 = np.array(route[i + 1]).reshape((-1, 2))
        i += 1
        nodes_distance = Euclidean_distance(xy_node1, xy_node2)
        list_of_distances.append(nodes_distance)

    path_distance = np.round(np.sum(list_of_distances), 2)
    streets_name = np.array(streets_name).astype(str)
    speed_limit = np.array(speed_limit).astype(int)
    return list_of_distances, streets_name, speed_limit
```

```python
## punkt rozpoczecia i zakończenia
ulica_start = 'UG'
ulica_koniec = 'molo'

source, target, W, lbl = Get_Start_End_Point(G, ulica_start, ulica_koniec)
```

```python
#Wyznaczenie najkrótszej trasy od punktu początkowego do punktu koncówego metodą  algorytmu Dijkstry.
route = nx.shortest_path(G, source = source, target = target, method ='dijkstra' )
```

### Opisanie atrybutów wyznaczonej trasy - sumarycznej odległości oraz jej przebiegu.

```python
list_of_distances, streets_name, speed_limit = Get_Path_Attributes(route)
```

```python
path_distance = np.round(np.sum(np.array(list_of_distances).reshape((-1)), axis = 0), 2)
```

```python
print(f"Najkrótszy dystans pieszej trasy z ulicy: {ulica_start} do {ulica_koniec} to {path_distance}m")
```

```python
print("Wyznaczona trasa przebiega kolejno przez ulice:")
trasa = [print(f" - {x}") for x in streets_name] 
```

### Wizualizacja Wyników
Wizualizacja trasy oraz grafu z układu ulic </br>

```python
fig, ax = plt.subplots(1, 1, figsize=(15,15),  sharex = True, sharey=True)

## Wyswietlenie całego grafu
pos =  dict(zip(G.nodes(), G.nodes()))
nx.draw_networkx_nodes(G, pos, node_size = 1,  ax = ax)
nx.draw_networkx_edges(G, pos, ax = ax, width = 1.5, edge_cmap= plt.cm.Blues)

#### Wyswietlenie przebiegu trasy
i = G.subgraph(route)
pos =  dict(zip(i.nodes(), i.nodes()))
nx.draw_networkx_edges(i, pos = pos, edge_color = 'r', width = 5.0 , ax = ax)
nx.draw_networkx_nodes(i, pos = pos, node_size = 10, node_color = 'r', ax = ax) #or even nx.draw(h,pos=pos,node_color='b') to get nodes and edges in one command

# Etykiety tekstwoe
ax.text(source[0], source[1], "Start",
        color = 'k', ha = 'center', 
        va = 'center', 
        fontdict = {"fontsize":'large',"fontweight":"bold"})

ax.text(target[0], target[1], "End", 
        color = 'k', ha = 'center', 
        va = 'center', 
        fontdict = {"fontsize":'large',"fontweight":"bold"})

plt.legend(["roads nodes", 'roads edges', 'path edges', 
            'path nodes', 'start node', 'end node'],
           title='Legenda', loc = 'upper right')

try:
    cx.add_basemap(ax, crs = "EPSG:2180")
except Exception as e:
    print("Brak połączenia z siecią")
    print(e)
    
plt.title(f"Przebieg trasy z {ulica_start} do {ulica_koniec}")
plt.show()
```
